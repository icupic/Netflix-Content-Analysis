```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python
df['month_id'] = df['Date'].dt.month
```

```python
df['day_id'] = df['Date'].dt.day
```

```python
df['year_id'] = df['Date'].dt.year
```

```python
df['BuildingArea'].fillna('NO AREA', inplace = True)
```

```python
#oglas postavljen 2017. ili 2018. godine

df[(df['year_id'] == 2017) | (df['year_id'] == 2018)] 
```

```python
#oglas postavljen između 2016 i 2019 godine
df[(df['year_id'] <= 2019) & (df['year_id'] >= 2016)]
```

```python
#srednja vrijednost
df['Rooms'].mean()
```

```python
# grupisana naselja u odnosu na  srednju vrijednost broja soba i broja parking mjesta nekretnine
df.groupby('Suburb', as_index=False).agg({"Rooms":'mean','Car':'mean'})
```

```python
#Grupisana naseljau odnosu na  srednju vrijednost broja soba i cijeni
df.groupby('Suburb', as_index=False).agg({"Rooms":'mean','Price':'mean'})
```

```python
#grupisana naselja u odnosu na srednju vrijednost cijene i udaljenosti
df.groupby('Suburb', as_index=False).agg({"Price":'mean','Distance':'mean'})
```

```python
# nekretnine čiji je broj soba između 5 i 7
df[(df['Rooms'] <= 7) & (df['Rooms'] >= 5)]
```

```python
#nekretnine koje se nalaze u naseljima Alton ili Melton
df[(df['Suburb'] == 'Altona') | (df['Suburb'] == 'Melton')]
```

```python
#Naselja čija je udaljenost između 5 i 20 km
df[(df['Distance'] <= 20) & (df['Distance'] >= 5)]
```

```python
#Nekretnine koje su izgrađene između 2000 i 2010 godine
df[(df['YearBuilt'] <= 2010) & (df['YearBuilt'] >= 2000)]
```

```python
#Grupisano po cijeni u odnosu na srednju vrijednost broja soba i udaljenosti naselja
df.groupby('Price', as_index=False).agg({"Rooms":'mean','Distance':'mean'})
```

```python
#Grupisano po cijeni u odnosu na srednju vrijednost vroja soba i imanja

df.groupby('Price', as_index=False).agg({"Rooms":'mean','Landsize':'mean'})
```

```python
#Grupisano po godini izgradnje u odnosu na srednju vrijednost broja soba i cijeni
df.groupby('YearBuilt', as_index=False).agg({"Rooms":'mean','Price':'mean'})
```

```python
#Grupisano po cijeni u odnosu na srednju vrijednost broja soba i broja kupatila
df.groupby('Price', as_index=False).agg({"Rooms":'mean','Bathroom':'mean'})
```

```python
#Nekretnine koje imaju između 3 i 5 kupatila
df[(df['Bathroom'] <= 5) & (df['Bathroom'] >= 3)]
```

```python
#Nekretnine koje imaju između 5 i 7 parking mjesta
df[(df['Car'] <= 7) & (df['Car'] >= 5)]
```

```python
#Nekretnine čija je cijena između 1.08e+06 i 	2.34e+06
df[(df['Price'] <= 	2.34e+06) & (df['Price'] >= 	1.08e+06)]
```

```python
Grupisan tip nekretnine po srednjoj vrijednosti broja soba i kupatila
df.groupby('Type', as_index=False).agg({"Rooms":'mean','Bathroom':'mean'})
```

```python
#Nekretnine koje imaju imanje između 150 i 200
df[(df['Landsize'] <= 200) & (df['Landsize'] >= 150)]
```

```python
#Grupisana srednja vrijednost broja soba i kupatila po broju soba nekretnine
df.groupby('Rooms', as_index=False).agg({"Price":'mean','Bathroom':'mean'})
```

```python
#Grupisana srednja cijena i broj kupatila u odnosu na godinu izgradnje
df.groupby('YearBuilt', as_index=False).agg({"Price":'mean','Bathroom':'mean'})
```
